```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
```

```python
requests.packages.urllib3.disable_warnings()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
page_url="https://books.toscrape.com/"
```

```python
page=requests.get(page_url)
```

```python
page
```

```python
raw_html=page.text
```

```python
raw_html
```

```python
print(raw_html)
```

```python
soup=BeautifulSoup(raw_html,"html.parser")
print(soup)
```

```python
page_ol=soup.select('ol')
```

```python
print(page_ol)
```

```python
page_h3=soup.select('ol > li > article > h3')#ol > li:nth-child(1) > article > h3
page_h3
```

```python
page_h3[0]
```

```python
page_h3[0].text
```

```python
len(page_h3)
```

```python
titles=[]
for i in page_h3:
  titles.append(i.text)
titles
```

```python
page_prices=soup.select("ol > li > article > div.product_price > p.price_color")
page_prices
```

```python
prices=[]
for i in page_prices:
  prices.append(i.text)
prices
```

```python
for i in range(0,len(prices)):
  prices[i]=prices[i].replace("l","" )
prices
```

```python
type(prices[0])
```

```python
import random
```

```python
matrix=[]
for i in range(0,len(prices)):
  node={}
  node["title"]=titles[i]
  node["price"]=prices[i]
  node['duration']=random.randint(60,180)
  node['raing']=random.randint(1,5)
  matrix.append(node)
matrix
```

```python
df=pd.DataFrame(matrix)
df
```
