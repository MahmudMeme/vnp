```python
page_wins=browser.find_elements(By.CSS_SELECTOR,'table > tbody > tr.team > td.wins')
```

```python
boston_wins=[]
for i in page_wins:
  boston_wins.append(i.text)
boston_wins
```

```python
page_losses=browser.find_elements(By.CSS_SELECTOR,'table > tbody > tr.team > td.losses')
```

```python
boston_losses=[]
for i in page_losses:
  boston_losses.append(i.text)
boston_losses
```

```python
page_years=browser.find_elements(By.CSS_SELECTOR,'table > tbody > tr.team > td.year')
```

```python
boston_years=[]
for i in page_years:
  boston_years.append(i.text)
boston_years
```

```python
for i in range(len(boston_losses)):
  boston_wins[i]=int(boston_wins[i])
  boston_losses[i]=int(boston_losses[i])
  boston_years[i]=int(boston_years[i])
```

```python
matrix_boston=[]
for i in range(len(boston_years)):
  node={}
  node['win']=boston_wins[i]
  node['lose']=boston_losses[i]
  node['year']=boston_years[i]
  matrix_boston.append(node)

df_boston=pd.DataFrame(matrix_boston)
df_boston
```
